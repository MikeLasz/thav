```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def vis_f1(n, d, C, title=True, graph="random"):
    #path = "C" + str(C) + "\\f1\\"
    if(graph == "random"):
        f1 = pd.read_csv( "../data/f1thresholded_C" + str(C) + "n" + str(n) + "d" + str(d) + ".csv", index_col="threshold")
        aux = pd.read_csv( "../data/aux_f1thresholded_C" +str(C) + "n" + str(n) + "d" + str(d) + ".csv")
    else:
        f1 = pd.read_csv( "../data/sff1thresholded_C" + str(C) + "n" + str(n) + "d" + str(d) + ".csv", index_col="threshold")
        aux = pd.read_csv( "../data/aux_sff1thresholded_C" +str(C) + "n" + str(n) + "d" + str(d) + ".csv")
    suggestedthreshold = 3 * C * aux.iloc[0,0]
    f1stars = aux.iloc[0, 1]
    f1ric = aux.iloc[0, 2]

    #plt.figure(figsize=(16,6)) This looks better, but does not fit goot into papers format
    plt.figure(figsize=(10,6))
    ax = sns.lineplot( data=f1["f1score"], label="f1-score")
    ax = sns.lineplot( data=f1["precision"], label="precision", linestyle="dashed")
    ax = sns.lineplot( data=f1["recall"], label="recall", linestyle="dotted")
    #ax.lines[1].set_linestyle("--", lw=5)
    ax.lines[0].set_linewidth(5)
    ax.lines[1].set(linestyle="--", linewidth=5)
    ax.lines[2].set(linestyle="dotted", linewidth=5)
    plt.axvline( suggestedthreshold, linewidth=5)
    plt.axhline( f1stars, label="stars", color="y", linestyle="dotted", linewidth=5)
    plt.axhline( f1ric, label="ric", color="green", linestyle="dashed", linewidth=5)

    
    leg = ax.legend()
    leg_lines = leg.get_lines()
    leg_lines[1].set_linestyle("dashed")
    leg_lines[2].set_linestyle("dotted")

    ax.set_ylabel('Score', fontsize=30)
    ax.set_ylim([0, 1.05])
    ax.set_xlabel("Threshold", fontsize=30)
    plt.legend(loc='upper right', fontsize=20)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    fig = ax.get_figure()
    fig.savefig("../plots/" + graph + "/n" + str(n) + "d" + str(d) + "f1C" + str(C) + ".png") 
    if title:
        plt.title("Performance of the thresholded AV, "+"C=" + str(C) + ", d=" + str(d) + ", n=" + str(n))
    plt.show()
```

# 1. n=300, d=200

## 1.1. random graph

```python
vis_f1(300, 200, 0.3, title=False)
vis_f1(300, 200, 0.2, title=False)
vis_f1(300, 200, 0.4, title=False)
```

## 1.2. scale-free graph

```python
vis_f1(300, 200, 0.3, title=False, graph="scale-free")
vis_f1(300, 200, 0.2, title=False, graph="scale-free")
vis_f1(300, 200, 0.4, title=False, graph="scale-free")
```
