```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def vis_C_detection(n, d, graph="random", show_legend=True, show_xlabel=True, show_C=True):    
    color = sns.color_palette()
    
    C_hat = pd.read_csv("../data/simu_results/C_detection/C_hatn" + str(n) + "d" + str(d) + graph + ".csv", index_col="r")
    
    
    
    F1 = pd.read_csv("../data/simu_results/C_detection/F1_Cn" + str(n) + "d" + str(d) + graph + ".csv", index_col="C")
    plt.figure(figsize=(8,6))
    
    #sns.set_theme(style="whitegrid")
    sns.lineplot(data=C_hat["est. C"])
    plt.show() 
    
    
    
    sns.lineplot(data=F1)
    plt.axvline()
    plt.ylim(0, 1)
    plt.show()
```

```python
vis_C_detection(300,200)
```
